### Outliers 

```python
numerical_cols=[col for col in df.columns if df[col].dtype == np.float64 or df[col].dtype == np.int64]
```

```python
numerical_cols
```

```python
for i,col in enumerate(numerical_cols):
    plt.figure(i)
    sns.boxplot(df[col])
```

```python
### bmi column contains outliers
def removeOutlier(col):
    sorted(col)
    quant1,quant2=col.quantile([0.25,0.75])
    IQR=quant2-quant1
    lowerRange=quant1-(1.5*IQR)
    upperRange=quant2+(1.5*IQR)
    return lowerRange,upperRange
```

```python
for col in numerical_cols:
    low,high=removeOutlier(df[col])
    df[col]=np.where(df[col]>high,high,df[col])
    df[col]=np.where(df[col]<low,low,df[col])
```

### Hyperparameter Tunning

```python
#Randomized Search CV

# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num = 12)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
from sklearn.model_selection import RandomizedSearchCV

# Random search of parameters, using 5 fold cross validation, 
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = rf_random, param_distributions = random_grid,scoring='neg_mean_squared_error', n_iter = 10, cv = 5, verbose=2, random_state=42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
prediction = rf_random.predict(X_test)
```

```python
metrics.r2_score(y_test, prediction)
```

```python
plt.figure(figsize = (8,8))
plt.scatter(y_test, prediction, alpha = 0.5)
plt.xlabel("y_test")
plt.ylabel("y_pred")
plt.show()
```

### Restaurant Recommendation

```python
df=pd.read_csv('ZomatoData.csv')
```

```python
df=df.dropna()
```

```python
df.shape
```

```python
df['comb_cuisines']=df['Cuisines'].dropna().map(lambda x:(''.join(x.split(',')).strip()))
df['comb_cuisines'].head()
```

```python
features=['comb_cuisines','Currency','Rating']
```

```python
def combined_features(row):
    return row['comb_cuisines']+" "+row['Currency']+" "+str(row['Rating'])
df["combined_features"] = df.apply(combined_features, axis =1)
```

```python
df["combined_features"].head(5)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
count_matrix = cv.fit_transform(df["combined_features"])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cosine_sim = cosine_similarity(count_matrix)
```

```python
movie_user_likes = "Rovereto"
def get_index_from_title(title):
    return df[df['Restaurant Name'] == title].index.values[0]
movie_index = get_index_from_title(movie_user_likes)
```

```python
similar_movies = list(enumerate(cosine_sim[movie_index]))
```

```python
sorted_similar_movies = sorted(similar_movies, key=lambda x:x[1], reverse=True)
```

```python
def get_title_from_index(index):
    return df[df.index == index]["Restaurant Name"].values[0]
i=0
for movie in sorted_similar_movies:
    print(get_title_from_index(movie[0]))
    i=i+1
    if i>5:
        break
```
