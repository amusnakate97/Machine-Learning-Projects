### Missing Values

```python
df.isnull().sum()
```

```python
sns.boxplot(df['arrival_delay_in_minutes'])
plt.show()
```

```python
### As we can see there are many outliers so we'll replace nan values with median
df['arrival_delay_in_minutes'].fillna(df['arrival_delay_in_minutes'].median(),inplace=True)             
```

### Outliers Detection and Removal

```python
numerical_columns=[col for col in df.columns if (df[col].dtype==np.float64 or df[col].dtype==np.int64)]
```

```python
numerical_columns
```

```python
for i,col in enumerate(numerical_columns):
    plt.figure(i)
    sns.boxplot(df[col])
    
```

```python
columns_with_outliers=['flight_distance','checkin_service','departure_delay_in_minutes','arrival_delay_in_minutes']
```

```python
def removeOutlier(col):
    sorted(col)
    quant1,quant2=col.quantile([0.25,0.75])
    IQR=quant2-quant1
    lowerRange=quant1-(1.5*IQR)
    upperRange=quant2+(1.5*IQR)
    return lowerRange,upperRange
```

```python
for col in columns_with_outliers:
    low,high=removeOutlier(df[col])
    df[col]=np.where(df[col]>high,high,df[col])
    df[col]=np.where(df[col]<low,low,df[col])
```

### Encoding Categorical Features

```python
cat_features=[col for col in df.columns if col not in numerical_columns]
cat_features
```

```python
df['Gender'].value_counts()
```

```python
Gender=pd.get_dummies(df['Gender'],drop_first=True)
```

```python
df['customer_type'].value_counts()
```

```python
customer_type=pd.get_dummies(df['customer_type'],drop_first=True)
```

```python
df['type_of_travel'].value_counts()
```

```python
type_of_travel=pd.get_dummies(df['type_of_travel'],drop_first=True)
```

```python
df['customer_class'].value_counts()
```

```python
### customer_class is ordinal 
class_dict={'Business':1,'Eco Plus':2,'Eco':3}
df['customer_class']=df['customer_class'].map(class_dict)
```

```python
df['satisfaction'].value_counts()
```

```python
satisfaction_dict={'neutral or dissatisfied':0,'satisfied':1}
df['satisfaction']=df['satisfaction'].map(satisfaction_dict)
```

```python
df.head()
```

```python
df.drop(['Gender','customer_type','type_of_travel'],axis=1,inplace=True)
```

```python
df=pd.concat([df,Gender,customer_type,type_of_travel],axis=1)
```

```python
df.head()
```

### Feature Selection

```python
plt.figure(figsize=(18,12))
sns.heatmap(df.corr(),annot=True)
```

```python
X=df.drop('satisfaction',axis=1)
y=df['satisfaction']
```

```python
from sklearn.ensemble import ExtraTreesRegressor
selection = ExtraTreesRegressor()
selection.fit(X, y)
```

```python
plt.figure(figsize = (12,8))
feat_importances = pd.Series(selection.feature_importances_, index=X.columns)
feat_importances.sort_values(ascending=False).plot(kind='barh')
plt.show()
```

```python
top_14=feat_importances.sort_values(ascending=False).nlargest(14).index
top_14
```

```python
df['satisfaction'].value_counts()
```

```python
X=df[top_20]
```

```python
X.columns
```

### Train-Test Data

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

### Model Building

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier()
```

```python
rf.fit(X_train,y_train)
```

```python
y_pred=rf.predict(X_test)
```

```python
from sklearn import metrics
```

```python
metrics.confusion_matrix(y_test,y_pred)
```

```python
metrics.accuracy_score(y_test,y_pred)
```

```python
import pickle
# open a file, where you ant to store the data
file = open('random_forest_classification_model.pkl', 'wb')

# dump information to that file
pickle.dump(rf, file)
```
